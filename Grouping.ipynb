#Randomization

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/MK-classroom/F2022/main/grouping.csv'
df = pd.read_csv(url, encoding = "CP949")
df
```

```python
Grouping = ["G1", "G2", "G3", "G4", "G5", "G6", "G7"]*3

import random
random.shuffle(Grouping)
Grouping

name = df["Name"]
d = {"Group": Grouping, "Members": name}
a = pd.DataFrame(data=d)
a.sort_values('Group')
```
